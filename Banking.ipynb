# Categorical Analysis

```python
# Converting numerical data into categorical data.taking an example as Estimated Income
bins = [0,100000,300000,float('inf')]
labels = ['low', 'Med', 'High']

df['Income Band'] = pd.cut(df["Estimated Income"],bins=bins, labels=labels, right= False)
df['Income Band'].value_counts().plot(kind='bar')
```

```python
categorical_cols = df[["BRId","GenderId","Amount of Credit Cards","Nationality","Occupation","Fee Structure", "Loyalty Classification","Properties Owned","Risk Weighting","IAId","Income Band"]]

for col in categorical_cols:
    print(f"Value counts for '{col}':")
    display(df[col].value_counts())
```

# Bivarient analysis

```python
for i, predictor in enumerate(df[["BRId","GenderId","Amount of Credit Cards","Nationality","Occupation","Fee Structure", "Loyalty Classification","Properties Owned","Risk Weighting","IAId","Income Band"]]):
    plt.figure(i)
    sns.countplot(data = df, x = predictor, hue = "Nationality")
```

```python
# Histplot of values for different Occupation
for col in categorical_cols:
    if col == "Occupation":
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.title("Histogram of " + col)
    plt.show()
```
